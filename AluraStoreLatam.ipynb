### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
tipo_variables = tienda1.dtypes
print(tipo_variables)

columnas_tienda1 = tienda1.columns
print(columnas_tienda1)
```

###**Ingresos totales por cada tienda**

```python
ventas_totales_tienda1 = tienda1['Precio'].sum()
print(f'Total de ventas en la Tienda 1: ${ventas_totales_tienda1:,.0f}\n')

ventas_totales_tienda2 = tienda2['Precio'].sum()
print(f'Total de ventas en la Tienda 2: ${ventas_totales_tienda2:,.0f}\n')

ventas_totales_tienda3 = tienda3['Precio'].sum()
print(f'Total de ventas en la Tienda 3: ${ventas_totales_tienda3:,.0f}\n')

ventas_totales_tienda4 = tienda4['Precio'].sum()
print(f'Total de ventas en la Tienda 4: ${ventas_totales_tienda4:,.0f}\n')

ventas_totales = {
    'Tienda 1': ventas_totales_tienda1,
    'Tienda 2': ventas_totales_tienda2,
    'Tienda 3': ventas_totales_tienda3,
    'Tienda 4': ventas_totales_tienda4
}

# Convertir a Serie de Pandas
serie_ventas_totales = pd.Series(ventas_totales)

serie_ventas_totales.plot(kind='bar', color='darkcyan', title='Ventas totales por tienda')
plt.ylabel('Ventas totales($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
venta_productos_por_categoria_tienda1 = tienda1.groupby('Categoría del Producto')['Producto'].count()
venta_productos_por_categoria_tienda1_ordenado = venta_productos_por_categoria_tienda1.sort_values(ascending=False)
print('Total de productos vendidos por categoría en la tienda 1:\n')
print(venta_productos_por_categoria_tienda1_ordenado.head(3),'\n')

venta_productos_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count()
venta_productos_por_categoria_tienda2_ordenado = venta_productos_por_categoria_tienda2.sort_values(ascending=False)
print('Total de productos vendidos por categoría en la tienda 2:\n')
print(venta_productos_por_categoria_tienda2_ordenado.head(3),'\n')

venta_productos_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count()
venta_productos_por_categoria_tienda3_ordenado = venta_productos_por_categoria_tienda3.sort_values(ascending=False)
print('Total de productos vendidos por categoría en la tienda 3:\n')
print(venta_productos_por_categoria_tienda3_ordenado.head(3),'\n')

venta_productos_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count()
venta_productos_por_categoria_tienda4_ordenado = venta_productos_por_categoria_tienda4.sort_values(ascending=False)
print('Total de productos vendidos por categoría en la tienda 4:\n')
print(venta_productos_por_categoria_tienda4_ordenado.head(3),'\n')
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda1['Calificación'].mean()
print(f'Calificación promedio de la tienda 1: {calificacion_promedio_tienda1:.2f}\n')

calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print(f'Calificación promedio de la tienda 2: {calificacion_promedio_tienda2:.2f}\n')

calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print(f'Calificación promedio de la tienda 3: {calificacion_promedio_tienda3:.2f}\n')

calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f'Calificación promedio de la tienda 4: {calificacion_promedio_tienda4:.2f}\n')
```

# 4. Productos más y menos vendidos

```python
# Productos más vendidos:
productos_mas_vendidos_tienda1 = tienda1['Producto'].value_counts(ascending=False)
print("Productos más vendidos tienda 1:\n")
print(productos_mas_vendidos_tienda1.head(3),'\n')

productos_mas_vendidos_tienda2 = tienda2['Producto'].value_counts(ascending=False)
print("Productos más vendidos tienda 2:\n")
print(productos_mas_vendidos_tienda2.head(3),'\n')

productos_mas_vendidos_tienda3 = tienda3['Producto'].value_counts(ascending=False)
print("Productos más vendidos tienda 3:\n")
print(productos_mas_vendidos_tienda3.head(3),'\n')

productos_mas_vendidos_tienda4 = tienda4['Producto'].value_counts(ascending=False)
print("Productos más vendidos tienda 4:\n")
print(productos_mas_vendidos_tienda4.head(3),'\n')

# Productos menos vendidos:
productos_menos_vendidos_tienda1 = tienda1['Producto'].value_counts(ascending=True)
print("Productos menos vendidos tienda 1:\n")
print(productos_menos_vendidos_tienda1.head(3),'\n')

productos_menos_vendidos_tienda2 = tienda2['Producto'].value_counts(ascending=True)
print("Productos menos vendidos tienda 2:\n")
print(productos_menos_vendidos_tienda2.head(3),'\n')

productos_menos_vendidos_tienda3 = tienda3['Producto'].value_counts(ascending=True)
print("Productos menos vendidos tienda 3:\n")
print(productos_menos_vendidos_tienda3.head(3),'\n')

productos_menos_vendidos_tienda4 = tienda4['Producto'].value_counts(ascending=True)
print("Productos menos vendidos tienda 4:\n")
print(productos_menos_vendidos_tienda4.head(3),'\n')
```

```python
from matplotlib import pyplot as plt

# Obtener los 3 productos más vendidos:
top3_productos_mas_vendidos_tienda1 = productos_mas_vendidos_tienda1.head(3)
top3_productos_mas_vendidos_tienda2 = productos_mas_vendidos_tienda2.head(3)
top3_productos_mas_vendidos_tienda3 = productos_mas_vendidos_tienda3.head(3)
top3_productos_mas_vendidos_tienda4 = productos_mas_vendidos_tienda4.head(3)

# Obtener los 3 productos menos vendidos:
top3_productos_menos_vendidos_tienda1 = productos_menos_vendidos_tienda1.head(3)
top3_productos_menos_vendidos_tienda2 = productos_menos_vendidos_tienda2.head(3)
top3_productos_menos_vendidos_tienda3 = productos_menos_vendidos_tienda3.head(3)
top3_productos_menos_vendidos_tienda4 = productos_menos_vendidos_tienda4.head(3)

# --- GRÁFICO 1: Top 3 productos más vendidos por tienda ---
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Top 3 productos MÁS vendidos por tienda', fontsize=16)

top_mas_vendidos = [
    top3_productos_mas_vendidos_tienda1,
    top3_productos_mas_vendidos_tienda2,
    top3_productos_mas_vendidos_tienda3,
    top3_productos_mas_vendidos_tienda4
]

titulos = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, ax in enumerate(axs.flat):
    top_mas_vendidos[i].plot(kind='bar', ax=ax, color='coral')
    ax.set_title(titulos[i])
    ax.set_ylabel('Cantidad de ventas')
    ax.set_xlabel('Producto')
    ax.set_xticklabels(top_mas_vendidos[i].index, rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# --- GRÁFICO 2: Top 3 productos MENOS vendidos por tienda ---
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Top 3 productos MENOS vendidos por tienda', fontsize=16)

top_menos_vendidos = [
    top3_productos_menos_vendidos_tienda1,
    top3_productos_menos_vendidos_tienda2,
    top3_productos_menos_vendidos_tienda3,
    top3_productos_menos_vendidos_tienda4
]

for i, ax in enumerate(axs.flat):
    top_menos_vendidos[i].plot(kind='bar', ax=ax, color='plum')
    ax.set_title(titulos[i])
    ax.set_ylabel('Cantidad de ventas')
    ax.set_xlabel('Producto')
    ax.set_xticklabels(top_menos_vendidos[i].index, rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

# 5. Envío promedio por tienda

```python
promedio_costo_envio_tienda1 = tienda1['Costo de envío'].mean()
print(f'El costo de envio promedio de la tienda 1 es de: ${promedio_costo_envio_tienda1:,.0f}\n')

promedio_costo_envio_tienda2 = tienda2['Costo de envío'].mean()
print(f'El costo de envio promedio de la tienda 2 es de: ${promedio_costo_envio_tienda2:,.0f}\n')

promedio_costo_envio_tienda3 = tienda3['Costo de envío'].mean()
print(f'El costo de envio promedio de la tienda 3 es de: ${promedio_costo_envio_tienda3:,.0f}\n')

promedio_costo_envio_tienda4 = tienda4['Costo de envío'].mean()
print(f'El costo de envio promedio de la tienda 4 es de: ${promedio_costo_envio_tienda4:,.0f}\n')
```

```python
costos_envio = {
    'Tienda 1': promedio_costo_envio_tienda1,
    'Tienda 2': promedio_costo_envio_tienda2,
    'Tienda 3': promedio_costo_envio_tienda3,
    'Tienda 4': promedio_costo_envio_tienda4
}

# Convertir a Serie de Pandas
serie_costos = pd.Series(costos_envio)

serie_costos.plot(kind='bar', color='skyblue', title='Costo de envío promedio por tienda')
plt.ylabel('Costo de envío ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
